Correlation between age and stamina

Correlation Matrix of Player Attributes

```python
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Player Attributes')
plt.show()
```

Strong Positive Correlations (Values close to 1):

Look for attributes with bright red colors. These indicate a strong positive relationship. For example, you might observe that strength has a high positive correlation with weight or height meaning that as one increases the other likely also increases.


Strong Negative Correlations (Values close to -1):

Look for attributes with dark blue colors. These show a strong inverse relationship. If you see one, it would indicate that as one variable increases, the other one decreases. (Note: based on your feature set, it is unlikely that you have strong negative correlation.)


Weak Correlations (Values close to 0):

Attributes with light or pale colors indicate weak or no linear correlation. These suggest that changes in one variable don't have a strong linear impact on the other.


Self Correlations (Values of 1):

The diagonal of the heatmap is always a perfect red, since each attribute is perfectly correlated with itself. This is not usually a point of interest.
